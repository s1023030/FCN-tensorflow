```python
import json
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.pyplot import imshow
import numpy as np
import cv2
%matplotlib inline
dir_path='mpii_human_pose_v1\images\\'
```

```python
raw_datas= open('data.json').readlines()
data=json.loads(raw_datas[10].strip())
```

```python
print(data['filename'])
img = cv2.imread(dir_path+data['filename'])
joints_pos=data['joint_pos']
for joint in joints_pos:
    x=int(joints_pos[str(joint)][0])
    y=int(joints_pos[str(joint)][1])
    cv2.circle(img,(x,y), 10, (0,0,255), -1)
plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
