```python
import json
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
import cv2
%matplotlib inline
dir_path='mpii_human_pose_v1\\output_images100x200\\'
img_height=200
img_width=100
```

```python
def draw_Gaussian_distribution(centerX,centerY,amplitude=1.0,stdx=15,stdy=15):
    stdx=stdx^2
    stdy=stdy^2
    x,y = np.meshgrid(np.linspace(0,img_width,img_width), np.linspace(0,img_height,img_height))
    x=x-centerX
    y=y-centerY
    x=x*x
    y=y*y
    x=x/(2*stdx)
    y=y/(2*stdy)
    combine=(x+y)*(-1)
    final=amplitude-amplitude*np.exp(combine)
    return final
```

```python
raw_datas= open('train_data\\new_data_100x200.json').readlines()
data=json.loads(raw_datas[10001].strip())
```

```python
print(data['filename'])
img = cv2.imread(dir_path+data['filename'])

gm_img_set=[]

joints_pos=data['joint_pos']
for joint in joints_pos:
    x=int(joints_pos[str(joint)][0])
    y=int(joints_pos[str(joint)][1])
    gm_img_set.append(draw_Gaussian_distribution(x,y))
    cv2.circle(img,(x,y), 1, (0,0,255), -1)

fig, axes=plt.subplots(5,4,figsize=(100,200))

times=0
gm_count=0
for ax in axes.flat:
    ax.axis("off")
    if times==0:
        ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    if times>3:
        ax.imshow(gm_img_set[gm_count],interpolation='lanczos')
        gm_count+=1
    times+=1
plt.show()
```
