```python
import tensorflow as tf
import numpy as np
import json
```

```python
learning_rate=0.001  #1e-5
training_iters=10000
display_step=20
batch_size=32
epoch=50

x=tf.placeholder(tf.float32,[None,1])
y=tf.placeholder(tf.float32,[None,1])
```

```python
w=tf.Variable(tf.random_normal(shape=[1]),name="w")
b=tf.Variable(tf.random_normal(shape=[1]),name="b")
```

```python
predict=x*w+b
cost=tf.reduce_mean((y-predict)*(y-predict))/batch_size
```

```python
optimizer = tf.train.RMSPropOptimizer(learning_rate=learning_rate, decay=0.0005, momentum=0.9, epsilon=1e-10)
gradients=optimizer.compute_gradients(cost,tf.trainable_variables())
def clipIfNotNone(grad):
    if grad is None:
        return grad
    return tf.clip_by_value(grad, -10, 10)
capped_gvs = [(clipIfNotNone(grad), var) for grad,var in gradients]
train_op = optimizer.apply_gradients(capped_gvs)
```

```python
saver=tf.train.Saver({"w":w,"b":b},max_to_keep=45)
init=tf.global_variables_initializer()
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
```

```python
def xy_next_batch(data_path='',batch_size=10,now_at=0):
    X=np.empty([batch_size,1])
    Y=np.empty([batch_size,1])
    raw_datas= open(data_path).readlines()
    for i in range(0,batch_size):
        data=json.loads(raw_datas[now_at+i].strip())
        xx=data['x']
        yy=data['y']
        X[i]=xx
        Y[i]=yy
    return X,Y
```

```python
with tf.Session(config=config) as sess:
    sess.run(init)
    for i in range(0,epoch):
        print("epoch:"+str(i)+"       !!!!!!!!!!!!")
        step=0
        now_at=0
        while step*batch_size<training_iters:
            tmpX,tmpY= xy_next_batch(data_path='practice\\practice.json',batch_size=10,now_at=0)
            sess.run(train_op,feed_dict = {x:tmpX,y:tmpY})
            loss=sess.run(cost,feed_dict = {x:tmpX,y:tmpY})
            if step%display_step==0:
                print("Iter "+str(step*batch_size)+", Minibatch Loss="+"{:.6f}".format(loss))
            step+=1
            now_at+=batch_size
        save_path = saver.save(sess, "practice/model/reg_"+str(i)+".ckpt")
print("done!")
```
